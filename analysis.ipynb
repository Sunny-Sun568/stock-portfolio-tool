```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("RBLX_ATVI.csv.csv")

# Split data by ticker
df_atvi = df[df['Ticker'] == 'ATVI'].copy()
df_rblx = df[df['Ticker'] == 'RBLX'].copy()

# Process date
df_atvi['DlyCalDt'] = pd.to_datetime(df_atvi['DlyCalDt'])
df_rblx['DlyCalDt'] = pd.to_datetime(df_rblx['DlyCalDt'])
df_atvi.set_index('DlyCalDt', inplace=True)
df_rblx.set_index('DlyCalDt', inplace=True)

# --- ATVI Charts ---
plt.figure(figsize=(12, 5))
plt.plot(df_atvi['DlyPrc'], color='blue')
plt.title('ATVI (Activision Blizzard) Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df_atvi['DlyRetx'], bins=40, color='blue', alpha=0.7)
plt.title('Distribution of ATVI Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# --- RBLX Charts ---
plt.figure(figsize=(12, 5))
plt.plot(df_rblx['DlyPrc'], color='red')
plt.title('RBLX (Roblox) Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df_rblx['DlyRetx'], bins=40, color='red', alpha=0.7)
plt.title('Distribution of RBLX Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# --- Comparison Charts ---
plt.figure(figsize=(12, 5))
plt.plot(df_atvi['DlyPrc'], label='ATVI (Activision Blizzard)', color='blue')
plt.plot(df_rblx['DlyPrc'], label='RBLX (Roblox)', color='red')
plt.title('Stock Price Trend Comparison: ATVI vs RBLX')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(9, 5))
plt.hist(df_atvi['DlyRetx'], bins=40, alpha=0.5, color='blue', label='ATVI')
plt.hist(df_rblx['DlyRetx'], bins=40, alpha=0.5, color='red', label='RBLX')
plt.title('Daily Return Distribution: ATVI vs RBLX')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
